```python
# in the directory path, you need to create a folder called /collections, for the files to be added.

from urllib.request import urlopen
from bs4 import BeautifulSoup
import datetime
import random
import re
from random import randint
from random import seed
import time

random.seed(datetime.datetime.now())

def getLinks(articleUrl):
    html=urlopen("http://en.wikipedia.org/"+articleUrl)
    bsObj=BeautifulSoup(html)
    findLinks=bsObj.find("div",{"id":"bodyContent"}).findAll("a",href=re.compile("^(/wiki/)((?!:).)*$"))
    findText=bsObj.find("div",{"id":"bodyContent"}).findAll('p')
    article=''
    for i in range(len(findText)):
        text = findText[i].get_text()
        text = re.sub(r'\[\d+\]', '', text) # regex to remove [x] i.e. references
        article+=text
        article=article.strip()
        articleUTF8=article.encode("UTF-8")

    try:
        text_file = open("C:/Users/Phu Wai Paing/Desktop/Course Books/Information Retrieval/data"+articleUrl[6:]+".txt", "wb") # EDIT YOUR DIRECTORY PATH HERE
        n = text_file.write(articleUTF8)
        text_file.close()
    except:
        print("skipped")
    return findLinks

articleUrl='/wiki/Yosemite' # just change accordingly.
    
links=getLinks(articleUrl)

while len(links)>0:
    newArticle = links[random.randint(0,len(links)-1)].attrs["href"]
    print(newArticle)
    try:
        links=getLinks(newArticle)
    except:
        print("ERROR!", newArticle)
        newArticle = links[random.randint(0,len(links)-1)].attrs["href"]
        #print(newArticle) # this is for verbose...
```

```python
import string
import nltk
from nltk import ngrams
```

```python
def extract_ngrams(data, num):
    n_grams = ngrams(nltk.word_tokenize(data), num)
    return [ ' '.join(grams) for grams in n_grams]
```

```python
with open('HillaryEmails/2.txt') as corpus:
    for line in corpus:
        # sequence = preprocess(line)
        preprocess_data = line.translate(str.maketrans('', '', string.punctuation))
        result = extract_ngrams(preprocess_data, 3)
        result = set(result)
        result = [x for x in set(result)]
        for r in result:
            print(r)
```
